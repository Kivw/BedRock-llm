```python
# 本项目使用的tokenizer是开源项目minimind中使用的tokenizer,其只有vocab_size仅为6400，在兼顾效率的同时满足绝大部分场景需求
# mimimind tokenizer 文件下载地址： https://github.com/jingyaogong/minimind/tree/master/model  （下载tokenizer.json和tokenizer.config.json即可）
```

```python
# load tokenizer
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("/home/jx/experiment/BedRock-llm/tokenizer")
tokenizer
```

```python
# 我们看到这个tokenizer中的special tokens 没有在instruction tuning中需要使用到的“system”和“user”等,我们需要手动添加。
# 一般来说预训练结束后也可以添加，但是此时需要更新模型的vocab_size,略显繁琐，所以我们这里先添加。值得注意的是我们在预训练阶段不会学习任何特殊token
special_tokens = ["system", "user", "assistant"]
tokenizer.add_special_tokens({"additional_special_tokens": special_tokens})
tokenizer
```

```python
# 我们可以看到我们成功添加了system、user、assistant三个特殊token,我们来验证一下是否符合预期
result = tokenizer.encode("system")
result # 记住此时的vocab_size为6403
```

```python
# 将修改后的tokenizer保存到本地
tokenizer.save_pretrained("/home/jx/experiment/BedRock-llm/tokenizer/tokenizer_modify")
```
